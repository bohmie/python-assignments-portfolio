```python
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point):
    def deriv(f, base_point):  # Estimate the derivative using symmetric approximation
        delta = 10**(-10)
        return (f(base_point + delta) - f(base_point - delta)) / (2 * delta)

    x_coords = [initial_point]  # Store x_n values
    y_coords = [f(initial_point)]  # Store f(x_n) values
    max_steps = 10000  # Maximum steps to avoid infinite loops

    for _ in range(max_steps):
        current_x = x_coords[-1]
        current_y = y_coords[-1]
        gradient = deriv(f, current_x)

        # Stop condition: when |f'(x)| is close to 0
        if abs(gradient) < 1e-10:
            break

        # Update step for gradient descent
        next_x = current_x - learning_rate * gradient
        next_y = f(next_x)

        x_coords.append(next_x)
        y_coords.append(next_y)

    # Plot the function and the descent path
    plot_range = np.linspace(min(x_coords) - 1, max(x_coords) + 1, 1000)
    function_range = [f(x) for x in plot_range]
    plt.plot(plot_range, function_range, label='Function f(x)')
    plt.scatter(x_coords, y_coords, color='red', label='Gradient Descent Path')
    plt.legend()
    plt.title('Gradient Descent Visualization')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.show()

    return round(x_coords[-1], 3), round(y_coords[-1], 3)  # Return final x and f(x)

# Test case with f(x) = x^2
result = gradient_descent(lambda x: x**2, learning_rate=0.8, initial_point=5)
print("Final x:", result[0], "Final f(x):", result[1])
```
