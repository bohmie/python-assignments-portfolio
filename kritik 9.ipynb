```python
#4a 
import sympy as sp

x, y = sp.symbols('x y')

# Define a function
f = sp.exp(x)* sp.sin(y) + y**3

# Differentiate f concerning x and y
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

# Differentiate f concerning x
print("the derivative of f with respect to x is:") 
df_dx
```

```python
#4a
# Differentiate f concerning y
print("the derivative of f with respect to y is:") 
df_dy  
```

```python
#4d
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
#ii. Define the function using SymPy:
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3 
#iii. Convert the SymPy expression to a numerical function:
j_func = sp.lambdify((x, y), j, 'numpy')
#iv. Create a grid of x and y values and evaluate j on this grid:
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)
#v. Plot the contour:
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x**3 - 3*x*y + y**3')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
